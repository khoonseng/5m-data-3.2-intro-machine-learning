# 1. Linear Regression Exercise:
   Using the California Housing dataset from scikit-learn, create a linear regression model to predict house prices.
   Evaluate the performance of Linear Regression on test set.

```python
from sklearn.datasets import fetch_california_housing
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
housing = fetch_california_housing(as_frame=True)
X = housing.data
y = housing.target

print(X)
print(y)
```

```python
# Split data 80-20 and reshape inputs for scikit-learn compatibility.
X_train, X_test, y_train, y_test = train_test_split(
                                   X,
                                   y, 
                                   test_size=0.2,
                                   random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Calculate Mean Squared Error (MSE) to quantify prediction accuracy (lower is better)
# Calculate R-squared (R2) to measure how well the independent variables explain the variance (closer to 1 is better)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R^2 Score         :", r2)
```

Based on the regression metrics defined in the sources, here is how to interpret the performance of this model on the test set:
1. Mean Squared Error (MSE): This represents the average squared difference between the actual house prices and the model's predictions.
- Interpretation: A lower MSE indicates the model's predictions are closer to the truth. Because the housing prices in this dataset are in units of $100,000, an MSE of, for example, 0.53 means the average squared error is roughly 0.53 units.
2. R-squared (R2): This represents the proportion of variance in house prices that is explained by the features (like average income or house age).
- Interpretation: This is a scale-invariant metric ranging from 0 to 1. An R2 of approximately 0.60 (typical for this dataset) would mean your model explains 60% of the variation in house prices, while the remaining 40% is due to "unpredictable factors" (ε) or noise.

# 2. Classification Exercise:
   Using the breast cancer dataset from scikit-learn, build classification models to predict malignant vs benign tumors.
   Compare Logistic Regression and KNN performance on test set.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

# 1. Load the dataset
# Features (X) are measurements; Label (y) is malignant/benign
cancer = load_breast_cancer()
X, y = cancer.data, cancer.target

# 2. Split the dataset into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Build Logistic Regression Model
# A supervised learning model that predicts categorical labels
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)

# 4. Build K-Nearest Neighbors (KNN) Model
# Classifies observations based on their proximity to 'K' neighbors
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)

# 5. Performance Evaluation
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))

print("\nLogistic Regression Report:\n", classification_report(y_test, y_pred_lr))
print("\nKNN Report:\n", classification_report(y_test, y_pred_knn))
```
